```python
# sales = pd.read_csv('/mnt/data/sales_monthly.csv')
sales = dataframes['sales_monthly.csv']
print(sales.head())

plt.figure(figsize=(10,6))
plt.plot(sales['month'], sales['sales'], marker='o', color='royalblue')

plt.title('Monthly Sales Over Time', fontsize=14)
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.xticks(rotation=45)

plt.tight_layout()
plt.savefig('a1_line.png', dpi=300)
plt.show()
```

•	A2. Plot Average sales per region using bar chart. Add value labels on bars. Export as **“/a2_bar.png” **

```python
avg_sales = sales.groupby('region')['sales'].mean().sort_values()

plt.figure(figsize=(8,6))
bars = plt.bar(avg_sales.index, avg_sales.values, color='teal')
plt.title('Average Sales per Region')
plt.xlabel('Region')
plt.ylabel('Average Sales ($)')

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
             f'{bar.get_height():.0f}', ha='center', va='bottom')

plt.tight_layout()
plt.savefig('a2_bar.png', dpi=300, bbox_inches='tight')
plt.show()
```

•	A.3 Distribution of “total_bill” from “tips_small.csv” with 15–30 bins. Include a mean line. Export as **“a3_hist.png” **

```python
tips = dataframes['tips_small.csv']

plt.figure(figsize=(8,6))
plt.hist(tips['total_bill'], bins=20, color='lightcoral', edgecolor='black')
mean_val = tips['total_bill'].mean()
plt.axvline(mean_val, color='blue', linestyle='--', label=f'Mean: {mean_val:.2f}')
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.savefig('a3_hist.png', dpi=300, bbox_inches='tight')
plt.show()
```

## Part B — Seaborn Statistical Plots

B1. Plot Mean Tip by day with 95% Confidence interval shown.  Export as **“b1_barPlot.png” **

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = pd.read_csv("tips_small.csv")

plt.figure(figsize=(8,6))
sns.barplot(data=tips, x="day", y="tip", ci=95, palette="coolwarm", capsize=0.1)

plt.title("Mean Tip by Day (95% CI)", fontsize=14)
plt.xlabel("Day of Week")
plt.ylabel("Mean Tip ($)")

plt.tight_layout()
plt.savefig("b1_barPlot.png", dpi=300, bbox_inches="tight")
plt.show()
```

B2. Plot violin plot of “total_bill” by “time” (Lunch vs Dinner) Export as **“b2_violin.png” **

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=tips, x="time", y="total_bill", palette="Set2")

plt.title("Distribution of Total Bill by Time", fontsize=14)
plt.xlabel("Meal Time")
plt.ylabel("Total Bill ($)")

plt.tight_layout()
plt.savefig("b2_violin.png", dpi=300, bbox_inches="tight")
plt.show()
```

B3.  Scatter Plot “bill_length_mm” vs “bill_depth_mm”, colored by “species”  Export as** “b3_scatter.png” **

```python
penguins = pd.read_csv("penguins_small.csv")

plt.figure(figsize=(8,6))
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    palette="viridis",
    s=80,
    edgecolor="white"
)

plt.title("Bill Length vs Bill Depth by Species", fontsize=14)
plt.xlabel("Bill Length (mm)")
plt.ylabel("Bill Depth (mm)")
plt.legend(title="Species")

plt.tight_layout()
plt.savefig("b3_scatter.png", dpi=300, bbox_inches="tight")
plt.show()
```

B4. Plot Correlation heatmap for the penguin’s dataset, using “viridis” heatmap.  Export as **“b4_heatmap.png” **

```python
num_cols = penguins.select_dtypes(include=["float", "int"])

plt.figure(figsize=(8,6))
sns.heatmap(num_cols.corr(), annot=True, cmap="viridis", fmt=".2f")

plt.title("Correlation Heatmap (Penguins Dataset)", fontsize=14)

plt.tight_layout()
plt.savefig("b4_heatmap.png", dpi=300, bbox_inches="tight")
plt.show()
```

## Part C — Plotly Interactivity

C1.  Plot an interactive scatter of “bill_length_mm” vs “bill_depth_mm” , color using species, Hover should show species and body mass.  **Export as “c1_scatter.html”**

```python
import plotly.express as px
penguins = pd.read_csv("penguins_small.csv")

fig_c1 = px.scatter(
    penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    color="species",
    hover_data=["species", "body_mass_g"],
    title="Bill Length vs Bill Depth by Species",
    template="plotly_white"
)

fig_c1.write_html("c1_scatter.html")

fig_c1.show()
```

C2.  Plot an interactive Bar Plot  Average tip % (tip / total_bill) by day with hover details. Add a hover template to clearly show % and counts. “Bonus” Export as **“c2_bar.html”**

```python
tips = pd.read_csv("tips_small.csv")

# Compute tip percentage
tips["tip_percent"] = 100 * tips["tip"] / tips["total_bill"]

avg_tips = (
    tips.groupby("day")
    .agg(avg_tip_percent=("tip_percent", "mean"), count=("tip_percent", "count"))
    .reset_index()
)

fig_c2 = px.bar(
    avg_tips,
    x="day",
    y="avg_tip_percent",
    hover_data={"avg_tip_percent": ":.2f", "count": True},
    text="avg_tip_percent",
    title="Average Tip Percentage by Day",
    template="plotly_white",
)

# Add hover template for cleaner info
fig_c2.update_traces(
    hovertemplate="<b>%{x}</b><br>Avg Tip %: %{y:.2f}%<br>Count: %{customdata[1]}<extra></extra>",
    texttemplate="%{y:.1f}%",
    textposition="outside"
)

fig_c2.update_layout(yaxis_title="Average Tip (%)")

# Export as HTML
fig_c2.write_html("c2_bar.html")

fig_c2.show()
```
