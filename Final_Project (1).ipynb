```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec
import os
from IPython import display as ipydisplay
```

```python
import seaborn as sns

corr_vars = [
    "Happiness_Index(1-10)",
    "Sleep_Quality(1-10)",
    "Exercise_Frequency(week)",
    "Daily_Screen_Time(hrs)"
]

corr_matrix = df[corr_vars].corr()

print("Correlation Matrix:")
print(corr_matrix)

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, vmin=-1, vmax=1, cmap="coolwarm")
plt.title("Correlation with Happiness Index")
plt.show()

happiness_corr = corr_matrix["Happiness_Index(1-10)"].drop("Happiness_Index(1-10)")
print("\nStrongest linear relationship with happiness:")
print(happiness_corr.abs().idxmax(), "=", happiness_corr.abs().max())
```

```python
group_means = df.groupby("Days_Without_Social_Media")[
    ["Stress_Level(1-10)", "Happiness_Index(1-10)"]
].mean()

print("\nMeans grouped by Days Without Social Media:")
print(group_means)

group_means.plot(kind="bar", figsize=(10,6))
plt.title("Stress & Happiness vs Days Without Social Media")
plt.ylabel("Average Score")
plt.xlabel("Days Without Social Media")
plt.show()
```

```python
if "Social_Media_Platform" in df.columns:

    platform_avg = df.groupby("Social_Media_Platform")["Happiness_Index(1-10)"].mean()
    print("\nAverage Happiness by Platform:")
    print(platform_avg)

    platform_avg.plot(kind="bar", figsize=(8,5))
    plt.title("Average Happiness Index by Platform")
    plt.ylabel("Average Happiness Index")
    plt.show()

else:
    print("\n No column named 'Social_Media_Platform' found in dataset.")
```

```python

optimal = df[
    (df["Happiness_Index(1-10)"] > 8) &
    (df["Stress_Level(1-10)"] < 3)
]

print("\nOptimal Profile Averages:")
print(optimal[
    ["Daily_Screen_Time(hrs)", "Sleep_Quality(1-10)", "Exercise_Frequency(week)"]
].mean())


avg_screen = optimal["Daily_Screen_Time(hrs)"].mean()
avg_sleep = optimal["Sleep_Quality(1-10)"].mean()
avg_exercise = optimal["Exercise_Frequency(week)"].mean()

print(f"""
Optimal Persona Summary:
People with very high happiness (>8) and very low stress (<3) tend to:
 Use social media only about {avg_screen:.1f} hours per day
 Maintain a sleep quality around {avg_sleep:.1f} / 10
 Exercise about {avg_exercise:.1f} times per week
""")
```

The analyses show that lifestyle patterns surrounding social media usage effect mental health and well-being. Individuals with longer breaks in screen time showed better sleep and less stress. However, overall screen time usage did not have a strong correlation with lower happiness, a slightly shocking result. Certain results ftom different social media platforms also showed that the usage of each platform was associated with different effects based on the purpose of the platform, arguing that some platforms may be more beneficial while others may be more harmful. However, the differences were not large enough to suggest limited usage of certain platforms over others. Nevertheless, the overall balancing of social media usage with outside factors like exercise and internet breaks and sleep did show a higher level of reorted happiness. Therefor, social media users should limit usage to 3 hours per day and maintain a sleep quality of 7.0/10.0 to optimize happiness index.
